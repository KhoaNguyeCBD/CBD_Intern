### Outlier Detection and Removal

Data is divided into four groups via the 25th, 50th and 75th values.

The IQR defines the middle 50% of the data, or the body of the data.

The IQR can be used to identify outliers by defining limits on the sample values that are a factor k of the IQR below the 25th percentile or above the 75th percentile. The common value for the factor k is the value 1.5. A factor k of 3 or more can be used to identify values that are extreme outliers or “far outs” when described in the context of box and whisker plots.

```python
data.describe()
```

The results show 8 numbers for each column in your original dataset:
- **count** shows how many rows have non-missing values.
- **mean** is the average. 
- **std** is the standard deviation which measures how numerically spread out the values are.
- **min, 25%, 50%, 75% and max** values, in each column from lowest to highest value. The first (smallest) value is the min. If you go a quarter way through the list, you'll find a number that is bigger than 25% of the values and smaller than 75% of the values. That is the 25% value (pronounced "25th percentile"). The 50th and 75th percentiles are defined analogously, and the max is the largest number.

### Convert Categorical Features by Using One-hot Coding

```python
#Seperate categorical from numerical data
categorical = dataNoNull.select_dtypes(exclude = [np.number])
numerical = dataNoNull.select_dtypes(include = [np.number])
categorical
```

```python
categorical.describe()
```

In Australia, Suburbs are contained within Regions and each Suburb has their own Postcode.
 - Drop the Postcode.
 
Suburb has 324 unique values and dealing with them takes lot of hard work while Regionname has only 8 unique value and One-hot encoding would suffice. 
 - Drop the Suburb.

Address(26732) will be dropted.

Only opt for features <= <font color='red'> 500 values</font>.

Date can be classified in year.

```python
categorical.drop(axis=1,columns=['Address','Postcode','Suburb'],inplace=True)
```

```python
def to_year(date_str):
    return datetime.datetime.strptime(date_str.strip(),'%d/%m/%Y').year
categorical['Date'] = categorical['Date'].apply(to_year)
```

```python
categorical['Date'].value_counts()
```

Only 3 value, so it can be accepted in one hot coding. Then convert these features as type of one hot encoding

```python
print(categorical.columns)
categorical_OneHot = pd.DataFrame()
for i in categorical:
    categorical_OneHot = pd.concat( [categorical_OneHot, pd.get_dummies(categorical[i], drop_first=True)], axis=1)
categorical_OneHot
```

**=> Actually, there are 19 columns, but in each feature 1 value is removed to ensure that it can avoid perfect multicollinearity.**
