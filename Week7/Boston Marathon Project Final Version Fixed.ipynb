### Outliers Detection and Removal 

Data is divided into four groups via the 25th, 50th and 75th values.

The IQR defines the middle 50% of the data, or the body of the data.

The IQR can be used to identify outliers by defining limits on the sample values that are a factor k of the IQR below the 25th percentile or above the 75th percentile. The common value for the factor k is the value 1.5. A factor k of 3 or more can be used to identify values that are extreme outliers or “far outs” when described in the context of box and whisker plots.

The results statistical description show:
- **count** shows how many rows have non-missing values.
- **mean** is the average. 
- **std** is the standard deviation which measures how numerically spread out the values are.
- **min, 25%, 50%, 75% and max** values, in each column from lowest to highest value. The first (smallest) value is the min. If you go a quarter way through the list, you'll find a number that is bigger than 25% of the values and smaller than 75% of the values. That is the 25% value (pronounced "25th percentile"). The 50th and 75th percentiles are defined analogously, and the max is the largest number.

## CLUSTERING NUMERICAL FEATURES

### Apply KMeans to Cluster Data 

Apply Kmean to cluster these dataset. Because KMeans require to determine k, Elbow method is used to choose the suitable k.

Inertia can be recognized as a measure of how internally coherent clusters are. It can be considerd as sum of distance between samples to their cluster. Therefore, the smalle interia, the better clustering. 

In this case, k ranges from 2 to 10 is checked.




```python
import time
ks = range(2,10)
inertias = []
tick = time.time()
for k in ks:
    kmeans = KMeans(n_clusters = k,random_state=0)
    kmeans.fit(data_PCA_StSc)
    inertias.append(kmeans.inertia_)
tock = time.time()
tock-tick
```

```python
inertias
```

```python
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

**=> Choose k = 3,4,5 will be considered.**

Tunning KMeans with 
- max_iter = [300,600,900]
- n_init = [10,20,30]
- ks = range(2,10)

```python
max_iter = [300,600,900]
n_init = [10,20,30]
ks = range(2,10)
inertias_tuning = []
inertias_st = []
tick = time.time()
for k in ks:
 for mi in max_iter:
  for nin in n_init:
    kmeans = KMeans(n_clusters=k, max_iter=mi, n_init=nin, random_state=0)
    kmeans.fit(data_PCA_StSc)
    inertias_tuning.append(kmeans.inertia_)
    inertias_st.append("k= " + str(k) + " max_iter= " + str(mi)+ " n_init= "+str(nin)+": " + str(kmeans.inertia_))
tock = time.time()
tock-tick
```

### Evaluate Result with Silhouette Coefficient

The Silhouette Coefficient is calculated using the mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. The Silhouette Coefficient for a sample is (b - a) / max(a, b). To clarify, b is the distance between a sample and the nearest cluster that the sample is not a part of. Note that Silhouette Coefficient is only defined if number of labels is 2 <= n_labels <= n_samples - 1.

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.

```python
ks = range(2,6)
import matplotlib.cm as cm
tick = time.time()
for n_clusters in ks:
    # Create a subplot with 1 row and 2 columns
    fig, ax1, = plt.subplots(1, 1)
    fig.set_size_inches(10, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(data_PCA_StSc) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = clusterer.fit_predict(data_PCA_StSc)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = metrics.silhouette_score(data_PCA_StSc, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = metrics.silhouette_samples(data_PCA_StSc, cluster_labels)
    labelSeri = pd.Series(cluster_labels)
    print(labelSeri.value_counts().sort_index()*100 /labelSeri.shape[0])
    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
tock = time.time()
plt.show()
tock-tick
```

```python
def CounterValuesFeature(group, feature, data, orderCount = False):
    '''
    group: string's name of lable
    feature: string's name of feature 
    '''
    newData = pd.DataFrame(index=np.sort(data[group].unique())) # set full index
    values = data[feature].unique()
    if(orderCount):
        values= np.sort(values)
    for i in values:
        #print(data[data[feature] == i].groupby(group)[feature].count())
        newData[i] = data[data[feature] == i].groupby(group)[feature].count().copy()
    newData.fillna(0,inplace=True)
    return newData
        
```

```python
def MeanValuesFeature(group, feature,yfeature, data, orderCount = False):
    '''
    group: string's name of lable
    feature: string's name of feature 
    '''
    newData = pd.DataFrame(index=np.sort(data[group].unique())) # set full index
    values = data[feature].unique()
    if(orderCount):
        values= np.sort(values)
    for i in values:
        #print(data[data[feature] == i].groupby(group)[feature].count())
        newData[i] = data[data[feature] == i].groupby(group)[yfeature].mean().copy()
    newData.fillna(0,inplace=True)
    return newData
        
```

### Evaluate Result with Silhouette Coefficient

The Silhouette Coefficient is calculated using the mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. The Silhouette Coefficient for a sample is (b - a) / max(a, b). To clarify, b is the distance between a sample and the nearest cluster that the sample is not a part of. Note that Silhouette Coefficient is only defined if number of labels is 2 <= n_labels <= n_samples - 1.

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more similar.



```python
tick = time.time()
silhouetteScore_OneHot = []
for k in ks_OneHot:
    kmeans_OneHot = KMeans(n_clusters = k,random_state=0)
    kmeans_OneHot.fit(data_PCA_StSc_OneHot)
    label_OneHot = kmeans_OneHot.predict(data_PCA_StSc_OneHot)
    silhouetteScore_OneHot.append(metrics.silhouette_score(data_PCA_StSc_OneHot, label_OneHot, metric='euclidean'))
tock = time.time()
tock-tick
```
